```python
Q1=np.percentile(df['longitude'],25,interpolation='midpoint')
Q2=np.percentile(df['longitude'],50,interpolation='midpoint')
Q3=np.percentile(df['longitude'],75,interpolation='midpoint')
print(Q1)
print(Q2)
print(Q3)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
print(upper_limit)
print(lower_limit)
```

```python
outlier=[]
for x in df['longitude']:
  if((x>upper_limit) or (x<lower_limit)):
    outlier.append(x)
outlier
```

# KNN

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
metric_k=[]
```

```python
neighbors=np.arange(3,15)
```

```python
for k in neighbors:
    knn=KNeighborsClassifier(n_neighbors=k)
    model_knn=knn.fit(x_train,y_train)
    y_pred_knn=model_knn.predict(x_test)
    acc=accuracy_score(y_test,y_pred_knn)
    metric_k.append(acc)
```

```python
metric_k
```

```python
plt.plot(neighbors,metric_k,'o--',color='y')
plt.xlabel('K Value')
plt.ylabel('Accuracy')
plt.grid()
```

```python
knn=KNeighborsClassifier(n_neighbors=14)
model_knn=knn.fit(x_train,y_train)
y_pred_knn=model_knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,precision_score,recall_score,f1_score,accuracy_score,classification_report
```
